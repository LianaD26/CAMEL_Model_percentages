## Extraer información de los csv

```python
import pandas as pd

def load_data(file_path):
    """
    Load data from a CSV file.

    Parameters:
    file_path (str): The path to the CSV file.

    Returns:
    pd.DataFrame: The loaded data as a DataFrame.
    """
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"Error loading data: {e}")
        return None


capital = load_data('../tablas/Capital.csv')
assests = load_data('../tablas/Assets.csv')
managerial = load_data('../tablas/Managerial.csv')
earnings = load_data('../tablas/Earnings.csv')
liquidity = load_data('../tablas/Liquidity.csv')
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
```

```python
df_long = pd.read_excel("../tablas/registros_categorizados.xlsx")
df_long = df_long[df_long["ano"] != 2014]
df_long.columns
```

```python
# lista de indicadores
indicadores = sorted(df_long["ID_indicador"].dropna().unique())

# graficar
def disp_plot_indicador(indicador):
    df_ind = df_long[df_long["ID_indicador"] == indicador]
    
    # dispersión ---
    fig, ax = plt.subplots()
    ax.scatter(df_ind["ano"], df_ind["valor"], alpha=0.5)

    plt.title(f"Distribución de {indicador}")
    plt.suptitle("")
    plt.xlabel("Año")
    plt.ylabel("Valor")
    plt.show()

def hist_plot_indicador(indicador):
    df_ind = df_long[df_long["ID_indicador"] == indicador]
    
    # histograma ---
    plt.hist(df_ind["valor"], bins=10, alpha=0.5)

    plt.title(f"Distribución de {indicador}")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.show()

def box_plot_indicador(indicador):
    df_ind = df_long[df_long["ID_indicador"] == indicador]
    
    # caja ---
    plt.figure()
    df_ind.boxplot(column="valor", by="ano")

    plt.title(f"Distribución de {indicador} por Año")
    plt.suptitle("")
    plt.xlabel("Año")
    plt.ylabel("Valor")
    plt.show()

# widget dropdown
dropdown = widgets.Dropdown(
    options=indicadores,
    description="Indicador:",
    layout=widgets.Layout(width="60%")
)
```

```python
# interacción
widgets.interactive(disp_plot_indicador, indicador=dropdown)
```

```python
widgets.interactive(hist_plot_indicador, indicador=dropdown)
```

```python
widgets.interactive(box_plot_indicador, indicador=dropdown)
```

```python
# lista de indicadores
indicadores = sorted(df_long["ID_indicador"].dropna().unique())

# lista de categorias
categorias = sorted(df_long["categoria"].dropna().unique())


# graficar
def disp_plot_indicador(indicador, categoria):

    df_ind = df_long[df_long["ID_indicador"] == indicador]

    if categoria != "Todas":
        df_ind = df_ind[df_ind["categoria"] == categoria]
    
    # dispersión ---
    fig, ax = plt.subplots()
    ax.scatter(df_ind["ano"], df_ind["valor"], alpha=0.5)

    plt.title(f"Distribución de {indicador} - {categoria}")
    plt.suptitle("")
    plt.xlabel("Año")
    plt.ylabel("Valor")
    plt.show()


def hist_plot_indicador(indicador, categoria):

    df_ind = df_long[df_long["ID_indicador"] == indicador]

    if categoria != "Todas":
        df_ind = df_ind[df_ind["categoria"] == categoria]
    
    # histograma ---
    plt.hist(df_ind["valor"], bins=10, alpha=0.5)

    plt.title(f"Distribución de {indicador} - {categoria}")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.show()


def box_plot_indicador(indicador, categoria):

    df_ind = df_long[df_long["ID_indicador"] == indicador]

    if categoria != "Todas":
        df_ind = df_ind[df_ind["categoria"] == categoria]
    
    # caja ---
    plt.figure()
    df_ind.boxplot(column="valor", by="ano")

    plt.title(f"Distribución de {indicador} por Año - {categoria}")
    plt.suptitle("")
    plt.xlabel("Año")
    plt.ylabel("Valor")
    plt.show()


# -----------------------------
# cálculo de rangos normales
# -----------------------------


# función general para widgets
def analizar(indicador, categoria):

    disp_plot_indicador(indicador, categoria)
    hist_plot_indicador(indicador, categoria)
    box_plot_indicador(indicador, categoria)


# -----------------------------
# widgets
# -----------------------------

dropdown_indicador = widgets.Dropdown(
    options=indicadores,
    description="Indicador:",
    layout=widgets.Layout(width="60%")
)

dropdown_categoria = widgets.Dropdown(
    options=["Todas"] + categorias,
    description="Categoría:",
    layout=widgets.Layout(width="60%")
)


widgets.interactive(
    analizar,
    indicador=dropdown_indicador,
    categoria=dropdown_categoria
)
```
