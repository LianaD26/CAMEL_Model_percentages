1. Extraer los valores para 110500, 111000, 111500 y la fecha para cada cooperativa

```python
import pandas as pd
coops = pd.read_excel("Datos_2013_2025_cooperativas.xlsx")
coops
```

```python
target = [110500, 111000, 111500]
df_filt = coops[coops["CUENTA"].isin(target)].copy()
```

```python
columns_base = ["CUENTA", "NOMBRE CUENTA", "fecha"]
coops_ = [col for col in coops.columns if col not in columns_base]
```

```python
df_activos_liquidos = (
    df_filt
    .groupby("fecha")[coops_]
    .sum()
    .reset_index()
)
```

```python
df_activos_liquidos = df_activos_liquidos[
    df_activos_liquidos["fecha"] >= "2020-01-01"
]
```

```python
df_activos_liquidos
```

```python
df_activos_liquidos.to_excel("activos_liquidos_coops.xlsx", index=False)
```

2. Extraer histórico del IRL

```python
irl_banks = pd.read_excel("Histórico_IRL - copia.xlsx")
irl_banks
```

```python
df_activos_liquidos["fecha_mes"] = df_activos_liquidos["fecha"].dt.to_period("M")
irl_banks["fecha_mes"] = irl_banks["Fecha"].dt.to_period("M")
```

```python
fechas_comunes = set(df_activos_liquidos["fecha_mes"]) & set(irl_banks["fecha_mes"])
```

```python
print(fechas_comunes)
```

```python
df_coop = df_activos_liquidos[
    df_activos_liquidos["fecha_mes"].isin(fechas_comunes)
].sort_values("fecha_mes")

df_bancos = irl_banks[
    irl_banks["fecha_mes"].isin(fechas_comunes)
].sort_values("fecha_mes")
```

```python
import numpy as np
import pandas as pd

# Filtrar desde marzo 2020
df_coop = df_coop[df_coop["fecha_mes"] >= "2020-03"]
df_bancos = df_bancos[df_bancos["fecha_mes"] >= "2020-03"]

# columnas cooperativas
cols_excluir = ["fecha", "fecha_mes"]
coop_nombres = [col for col in df_coop.columns if col not in cols_excluir]

# asegurar numéricos
df_coop[coop_nombres] = df_coop[coop_nombres].apply(pd.to_numeric, errors="coerce")
df_bancos["ALM PromedioMes"] = pd.to_numeric(df_bancos["ALM PromedioMes"], errors="coerce")

resultados = []

# iterar por fecha
for fecha in df_coop["fecha_mes"].unique():

    df_coop_mes = df_coop[df_coop["fecha_mes"] == fecha]
    df_bancos_mes = df_bancos[df_bancos["fecha_mes"] == fecha]

    # valores de bancos en ese mes
    bancos_mes = df_bancos_mes[["Nombre Entidad", "ALM PromedioMes"]].dropna()

    for coop in coop_nombres:
        valor_coop = df_coop_mes[coop].values[0]

        if np.isnan(valor_coop):
            continue

        mejor_banco = None
        mejor_distancia = np.inf
        mejor_valor_banco = None

        # comparar contra TODOS los bancos de ese mes
        for _, row in bancos_mes.iterrows():
            banco = row["Nombre Entidad"]
            valor_banco = row["ALM PromedioMes"]

            distancia = abs(valor_coop - valor_banco)

            if distancia < mejor_distancia:
                mejor_distancia = distancia
                mejor_banco = banco
                mejor_valor_banco = valor_banco

        resultados.append({
            "fecha_mes": fecha,
            "cooperativa": coop,
            "valor_coop": valor_coop,
            "banco_mas_parecido": mejor_banco,
            "valor_banco": mejor_valor_banco,
            "distancia": mejor_distancia
        })

df_matching = pd.DataFrame(resultados)
```
