```python
import pandas as pd
```

```python
df_coperativas = pd.read_csv("../../tablas/camel/Registros_CAMEL.csv")
df_coperativas
```

```python
df_irl = pd.read_excel("../../tablas/df_matching_coop_bancos_IRL.xlsx")
df_irl
```

```python
df_irl["fecha_mes"] = df_irl["fecha_mes"].astype(str)

df_irl["ano"] = df_irl["fecha_mes"].str[:4].astype(int)
df_irl["mes"] = df_irl["fecha_mes"].str[5:7].astype(int)
```

```python
df_coop_irl = df_coperativas[
    df_coperativas["ID_indicador"] == "Indicador de Riesgo de Liquidez - IRL"
].copy()
```

```python
df_merge = pd.merge(
    df_coop_irl,
    df_irl[["cooperativa", "ano", "mes", "IRL"]],
    left_on=["ID_cooperativa", "ano", "mes"],
    right_on=["cooperativa", "ano", "mes"],
    how="inner"   
)
```

```python
df_merge = df_merge.drop(["cooperativa", "valor"], axis=1)
```

```python
df_merge = df_merge.rename(columns={"IRL": "valor"})
```

```python
df_merge
```

Concatenar tablas

```python
df_total = pd.concat([df_coperativas, df_merge], ignore_index=True)
```

```python
df_total = df_total.drop_duplicates(
    subset=["ano", "mes", "ID_indicador", "ID_cooperativa"],
    keep="last"
)
```

```python
df_total
```

```python
df_total.to_excel("../../tablas/camel/Registros_CAMEL_con_IRL.xlsx", index=False)
```
