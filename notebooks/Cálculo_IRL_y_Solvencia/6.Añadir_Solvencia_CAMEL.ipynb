```python
import pandas as pd
```

```python
data_solvencia = pd.read_excel("../../tablas/registros_solvencia.xlsx")
data_solvencia
```

```python
data_solvencia_bancos = data_solvencia[data_solvencia["Concepto"]=="10. Relación solvencia básica (%)"]
data_solvencia_bancos
```

```python
data_solvencia_bancos = data_solvencia_bancos.drop(columns=["Tipo entidad", "Código entidad", "Concepto"])
data_solvencia_bancos
```

## Concatenar estos valores con los de IRL

```python
# leer tabla completa del IRL
data = pd.read_excel("../../tablas/camel/Registros_CAMEL_con_IRL.xlsx")
data
```

```python
data_matching = pd.read_excel("../../tablas/df_matching_coop_bancos_IRL.xlsx")
data_matching
```

```python
# Convertir el formato de fecha de data_solvencia_bancos a formato YYYY-MM
data_solvencia_bancos['Fecha'] = pd.to_datetime(data_solvencia_bancos['Fecha'])
data_solvencia_bancos['fecha_mes'] = data_solvencia_bancos['Fecha'].dt.strftime('%Y-%m')

# Renombrar la columna Entidad a Nombre Entidad para que coincida con data_matching
data_solvencia_bancos_renamed = data_solvencia_bancos[['fecha_mes', 'Entidad', 'Valor']].copy()
data_solvencia_bancos_renamed = data_solvencia_bancos_renamed.rename(columns={'Entidad': 'Nombre Entidad', 'Valor': 'Solvencia'})

# Hacer merge de los dataframes donde coincidan la fecha y el nombre del banco
data_matching_con_solvencia = data_matching.merge(
    data_solvencia_bancos_renamed,
    on=['fecha_mes', 'Nombre Entidad'],
    how='left'
)

data_matching_con_solvencia
```

```python
# Reemplazar valores NaN con 0.0 en la columna Solvencia
data_matching_con_solvencia['Solvencia'] = data_matching_con_solvencia['Solvencia'].fillna(0.0)

# Verificar el resultado
print("Verificación final:")
print(f"Valores no nulos: {data_matching_con_solvencia['Solvencia'].notna().sum()}")
print(f"Valores que son 0.0: {(data_matching_con_solvencia['Solvencia'] == 0.0).sum()}")
print(f"\nMuestra de resultados:")
print(data_matching_con_solvencia[['fecha_mes', 'Nombre Entidad', 'Solvencia']].head(20))
```

## Concatenar con la tabla general de todos los indicadores

```python
df_coperativas = pd.read_excel("../../tablas/camel/Registros_CAMEL_con_IRL.xlsx")
df_coperativas
```

```python
data_matching_con_solvencia["fecha_mes"] = data_matching_con_solvencia["fecha_mes"].astype(str)

data_matching_con_solvencia["ano"] = data_matching_con_solvencia["fecha_mes"].str[:4].astype(int)
data_matching_con_solvencia["mes"] = data_matching_con_solvencia["fecha_mes"].str[5:7].astype(int)
```

```python
df_coop_solv = df_coperativas[
    df_coperativas["ID_indicador"] == "Relación Solvencia"
].copy()
```

```python
df_merge = pd.merge(
    df_coop_solv,
    data_matching_con_solvencia[["cooperativa", "ano", "mes", "Solvencia"]],
    left_on=["ID_cooperativa", "ano", "mes"],
    right_on=["cooperativa", "ano", "mes"],
    how="inner"   
)
```

```python
df_merge = df_merge.drop(["cooperativa", "valor"], axis=1)
```

```python
df_merge = df_merge.rename(columns={"Solvencia": "valor"})
```

```python
df_merge
```

```python
df_total = pd.concat([df_coperativas, df_merge], ignore_index=True)
```

```python
df_total = df_total.drop_duplicates(
    subset=["ano", "mes", "ID_indicador", "ID_cooperativa"],
    keep="last"
)
```
