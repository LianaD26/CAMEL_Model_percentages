## Extraer información de los csv

```python
import pandas as pd

def load_data(file_path):
    """
    Load data from a CSV file.

    Parameters:
    file_path (str): The path to the CSV file.

    Returns:
    pd.DataFrame: The loaded data as a DataFrame.
    """
    try:
        data = pd.read_csv(file_path)
        return data
    except Exception as e:
        print(f"Error loading data: {e}")
        return None


capital = load_data('/home/lia/Documentos/repos/CAMEL_Model_Percentages/CAMEL_Model_percentages/tablas/Capital.csv')
assests = load_data('/home/lia/Documentos/repos/CAMEL_Model_Percentages/CAMEL_Model_percentages/tablas/Assets.csv')
managerial = load_data('/home/lia/Documentos/repos/CAMEL_Model_Percentages/CAMEL_Model_percentages/tablas/Managerial.csv')
earnings = load_data('/home/lia/Documentos/repos/CAMEL_Model_Percentages/CAMEL_Model_percentages/tablas/Earnings.csv')
liquidity = load_data('/home/lia/Documentos/repos/CAMEL_Model_Percentages/CAMEL_Model_percentages/tablas/Liquidity.csv')
```

### Cambiar el formato de las tablas para graficar mejor

```python
df = pd.concat([capital, assests, managerial, earnings, liquidity], axis=0, ignore_index=True)
df.head()
```

```python
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display
```

```python
cols_fijas = ["Año", "Mes", "índice CAMEL"]
cols_coops = [c for c in df.columns if c not in cols_fijas]

df_long = df.melt(
    id_vars=cols_fijas,
    value_vars=cols_coops,
    var_name="cooperativa",
    value_name="valor"
)

# lista de indicadores
indicadores = sorted(df_long["índice CAMEL"].dropna().unique())

# graficar
def disp_plot_indicador(indicador):
    df_ind = df_long[df_long["índice CAMEL"] == indicador]
    
    # dispersión ---
    fig, ax = plt.subplots()
    ax.scatter(df_ind["Año"], df_ind["valor"], alpha=0.5)

    plt.title(f"Distribución de {indicador}")
    plt.suptitle("")
    plt.xlabel("Año")
    plt.ylabel("Valor")
    plt.show()

def hist_plot_indicador(indicador):
    df_ind = df_long[df_long["índice CAMEL"] == indicador]
    
    # histograma ---
    plt.hist(df_ind["valor"], bins=10, alpha=0.5)

    plt.title(f"Distribución de {indicador}")
    plt.xlabel("Valor")
    plt.ylabel("Frecuencia")
    plt.show()

def box_plot_indicador(indicador):
    df_ind = df_long[df_long["índice CAMEL"] == indicador]
    
    # caja ---
    plt.figure()
    df_ind.boxplot(column="valor", by="Año")

    plt.title(f"Distribución de {indicador} por Año")
    plt.suptitle("")
    plt.xlabel("Año")
    plt.ylabel("Valor")
    plt.show()

# widget dropdown
dropdown = widgets.Dropdown(
    options=indicadores,
    description="Indicador:",
    layout=widgets.Layout(width="60%")
)
```

```python
# interacción
widgets.interactive(disp_plot_indicador, indicador=dropdown)
```

```python
widgets.interactive(hist_plot_indicador, indicador=dropdown)
```

```python
widgets.interactive(box_plot_indicador, indicador=dropdown)
```
