# Modelo CAMEL

Usar los csv generados para los indicadores de cada variable y concatenarlos en una misma tabla de registros

```python
import pandas as pd

capital = pd.read_csv("../../tablas/camel/Capital.csv")
assets = pd.read_csv("../../tablas/camel/Assets.csv")
managerial = pd.read_csv("../../tablas/camel/Managerial.csv")
earnings = pd.read_csv("../../tablas/camel/Earnings.csv")
liquidity = pd.read_csv("../../tablas/camel/Liquidity.csv")

registros_camel = pd.concat([capital, assets, managerial, earnings, liquidity], ignore_index=True)
registros_camel.head(4)
```

Ajustar la estructura de la tabla (mejor)

```python
# Convertir de formato ancho a largo
registros_camel_long = registros_camel.melt(
    id_vars=['Año', 'Mes', 'índice CAMEL'],
    var_name='ID_cooperativa',
    value_name='valor'
)

# Renombrar columnas
registros_camel_long = registros_camel_long.rename(columns={
    'Año': 'ano',
    'Mes': 'mes',
    'índice CAMEL': 'ID_indicador'
})

# Crear ID_registro
registros_camel_long.insert(
    0,
    'ID_registro',
    range(1, len(registros_camel_long) + 1)
)

registros_camel_long.head()
```

```python
registros_camel_long.to_csv("../../tablas/camel/Registros_CAMEL.csv", index=False, encoding="utf-8")
```
