```python
# revisar los indicadores de Earnings
import pandas as pd

data = pd.read_excel("../tablas/camel/Registros_CAMEL_con_IRL.xlsx")
data.head(3)
```

```python
# Extraer los indicadores de Earnings
earnings = pd.read_csv("../tablas/camel/Earnings.csv")
earnings.head(3)
```

```python
# Ajustar estructura de la tabla
# Convertir de formato ancho a largo
registros_earnings = earnings.melt(
    id_vars=['Año', 'Mes', 'índice CAMEL'],
    var_name='ID_cooperativa',
    value_name='valor'
)

# Renombrar columnas
registros_earnings = registros_earnings.rename(columns={
    'Año': 'ano',
    'Mes': 'mes',
    'índice CAMEL': 'ID_indicador'
})

# Crear ID_registro
registros_earnings.insert(
    0,
    'ID_registro',
    range(1, len(registros_earnings) + 1)
)

registros_earnings.head()
```

```python
# Concatenar con la tabla de registros CAMEL
import pandas as pd

# df1 y df2 tienen las mismas columnas
df_combined = pd.concat([data, registros_earnings], ignore_index=True)

df_result = df_combined.drop_duplicates(
    subset=['ano', 'mes', 'ID_indicador', 'ID_cooperativa'],
    keep='last'
)
```

```python
df_result.to_excel("../tablas/camel/Registros_CAMEL_con_IRL.xlsx", index=False)
```
