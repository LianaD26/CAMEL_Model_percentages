```python
import pandas as pd
import numpy as np
```

```python
datos_viejos=pd.read_excel("../tablas/registros_solvencia_actualizado.xlsx")
datos_viejos=datos_viejos.drop(columns=["Unnamed: 0"])
datos_viejos=datos_viejos[datos_viejos["ID_indicador"]!='Indicador de Riesgo de Liquidez - IRL']
datos_viejos.columns
```

# transformar datos

```python
datos_activos_liquidos=pd.read_excel("../tablas/activos_liquidos_coops.xlsx")
datos_activos_liquidos.head()
```

```python
datos_activos_liquidos_melted = datos_activos_liquidos.melt(id_vars=["fecha"], 
                                                            value_vars=datos_activos_liquidos.columns[1:].tolist(),
                                                            var_name="ID_cooperativa", 
                                                            value_name="valor")
```

```python
datos_activos_liquidos_melted
```

```python
datos_activos_liquidos_melted["ID_registro"]=99
datos_activos_liquidos_melted["fecha"]=pd.to_datetime(datos_activos_liquidos_melted["fecha"], format="%Y-%m-%d")
datos_activos_liquidos_melted["ano"]=datos_activos_liquidos_melted["fecha"].dt.year
datos_activos_liquidos_melted["mes"]=datos_activos_liquidos_melted["fecha"].dt.month
datos_activos_liquidos_melted["ID_indicador"]="IRL"
datos_activos_liquidos_melted=datos_activos_liquidos_melted.drop(columns=["fecha"])
datos_activos_liquidos_melted.head()
```

# unir datos, teniendo en cuenta que fecha minima 2020 y fecha maxima 2025

```python
datos_viejos_mescla=datos_viejos.loc[(datos_viejos["ano"]>=2020)&(datos_viejos["ano"]<=2025)]
datos_activos_liquidos_mescla=datos_activos_liquidos_melted.loc[(datos_activos_liquidos_melted["ano"]>=2020)&(datos_activos_liquidos_melted["ano"]<=2025)]
datos_concatenados=pd.concat([datos_viejos_mescla,datos_activos_liquidos_mescla], ignore_index=True)
datos_concatenados.head()
```

```python
datos_concatenados.to_excel("../tablas/registros_irl_para_pca.xlsx", index=False)
```
